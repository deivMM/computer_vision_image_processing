```python
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from glob import glob
```

```python
img_files = glob('images/*jpg')
img_files
```

```python
img_mpl = plt.imread('images/imagen_2.jpg')
img_cv2 = cv2.imread('images/imagen_2.jpg')

print(f'El shape es: {img_cv2.shape} | Altura: {img_cv2.shape[0]} | Anchura: {img_cv2.shape[1]} | Canales: {img_cv2.shape[2]}')
```

```python
f, ax = plt.subplots(figsize=(8, 8))
ax.imshow(img_mpl)
ax.axis('off')
plt.show()
```

## cv2 reads in channels BGR and matplotlib in RBG

```python
f, axs = plt.subplots(1, 3, figsize=(15,5))

axs[0].imshow(img_mpl[:,:,0], cmap='Reds')
axs[1].imshow(img_mpl[:,:,1], cmap='Greens')
axs[2].imshow(img_mpl[:,:,2], cmap='Blues')

axs[0].axis('off')
axs[1].axis('off')
axs[2].axis('off')
axs[0].set_title('Red channel')
axs[1].set_title('Green channel')
axs[2].set_title('Blue channel')
plt.show()


# f, axs = plt.subplots(1, 3, figsize=(15,5))
# axs[0].imshow(img_cv2[:,:,0], cmap='Reds')
# axs[1].imshow(img_cv2[:,:,1], cmap='Greens')
# axs[2].imshow(img_cv2[:,:,2], cmap='Blues')

# axs[0].axis('off')
# axs[1].axis('off')
# axs[2].axis('off')
# axs[0].set_title('Red channel')
# axs[1].set_title('Green channel')
# axs[2].set_title('Blue channel')
# plt.show()
```

```python
f, ax = plt.subplots(figsize=(6,6))
ax.imshow(img_cv2)
ax.axis('off')
plt.show()

img_cv2_rgb = cv2.cvtColor(img_cv2, cv2.COLOR_BGR2RGB)

f, ax = plt.subplots(figsize=(6,6))
ax.imshow(img_cv2_rgb)
ax.axis('off')
plt.show()
```

```python
img_gray = cv2.cvtColor(img_cv2, cv2.COLOR_BGR2GRAY)

f, ax = plt.subplots(figsize=(8, 8))
ax.imshow(img_gray, cmap ='gray')
ax.axis('off')
plt.show()
```

## resize an image

```python
resized_image = cv2.resize(img_cv2, None, fx = .01, fy=.01)
f, ax = plt.subplots(figsize=(6,6))
ax.imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB))
ax.axis('off')
plt.show()
```

```python
resized_image = cv2.resize(img_cv2, (100,100))
f, ax = plt.subplots(figsize=(6,6))
ax.imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB))
ax.axis('off')
plt.show()  
```

```python
resized_image = cv2.resize(img_cv2, (5000,5000), interpolation = cv2.INTER_CUBIC)
f, ax = plt.subplots(figsize=(6,6))
ax.imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB))
ax.axis('off')
plt.show()
resized_image.shape
```

```python
kernel_sharpening = np.array([[-1,-1,-1], 
                              [-1,9,-1], 
                              [-1,-1,-1]])

kernel_sharpening = np.array([[-1,-1,-1], 
                              [-1,8,-1], 
                              [-1,-1,-1]])


sharpened = cv2.filter2D(img_mpl, -1, kernel_sharpening)

fig, ax = plt.subplots(figsize=(8, 8))
ax.imshow(sharpened)
ax.axis('off')
ax.set_title('Sharpened Image')
plt.show()
```
