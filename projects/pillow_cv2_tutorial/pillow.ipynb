```python
from PIL import Image, ImageChops, ImageFilter
```

```python
im1 = Image.open('images/imagen_1.jpg')
print(f'size of the image in pixels: {im1.size} | colour mode: {im1.mode}' )
im1
```

```python
resize_factor = .5
ancho, alto = im1.size
print(im1.size)
imagen_redimensionada = im1.resize((int(ancho//(1/resize_factor)), int(alto//(1/resize_factor))))
print(imagen_redimensionada.size)
imagen_redimensionada
```

```python
img1_copia = im1.copy()
max_size = (300, 300)
print(im1.size)
img1_copia.thumbnail(max_size)
print(img1_copia.size)
img1_copia
```

```python
img1_copia.rotate(90)
```

```python
img1_copia.convert('L')
```

```python
img1_copia.filter(ImageFilter.GaussianBlur(10))
```

```python
img1_copia.filter(ImageFilter.FIND_EDGES())
```
